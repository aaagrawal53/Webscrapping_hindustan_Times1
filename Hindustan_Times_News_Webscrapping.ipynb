# Extracting news URL

```python
url=[]
for link in soup.find_all("div", attrs={'data-weburl': re.compile("^https://www.hindustantimes.com/entertainment/")}):
    url.append(link.get('data-weburl'))
```

# Creating dataframe with possible variables

```python
import pandas as pd
def news(soup):
    title = [i.text.strip() for i in soup.find_all('h3',class_='hdg3')]
    pub = [i.find_all('div')[1].text.strip() for i in soup.select('div.actionDiv.flexElm.topTime')]
    desc=[i.text.strip() for i in soup.find_all("h2", class_="sortDec")]
    return pd.DataFrame({"Title":title,'Published Date':pub,"Description":desc})
```

```python
df = news(soup)
df
```

```python
df['URL']=url
df.head()
```
